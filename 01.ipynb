```python
#mengenal object oriented style untuk melakukan plotting
x= np.linspace(0, 2, 100)
x
```

```python
#membuat sebuah figure dan sebuah axes
fig, ax = plt.subplots() 

#plotting tiga variant data pada axes
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')

#menyertakan x-label pada axes
ax.set_xlabel('x label')

#menyertakan y-label pada axes
ax.set_ylabel('y label')

#menyertakan title pada axes
ax.set_title('simple plot')

#menyertakan legend
ax.legend()
```
