```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
#Sample Data
#Nilai Mean dan Nilai Standard Deviation
mu, sigma = 100, 15

#Normal Distribution
x = mu + sigma * np.random.randn(10000)
x                               
```

```python
x.shape
```

```python
#Histogram Dengan Pyplot Style
plt.hist(x,
         bins = 50,
         facecolor = 'g',
         alpha = 0.75)

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Histogram')

plt.text(45, 500, r'$\mu=100,\ \sigma=15$')
plt.grid()

plt.show()
```

```python
#Histogram Dengan 00 Sytle
fig, ax = plt.subplots()

ax.hist(x,
        bins = 50,
        facecolor = 'r',
        alpha = 0.75)

ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')
ax.set_title('Contoh Histogram')

ax.text(45, 500, r'$\mu=100,\ \sigma=15$')
ax.grid()

plt.show()
```
